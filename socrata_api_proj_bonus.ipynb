1. Adjust your request so that it returns data just for the United States.

```python
# Indicator: GDP per capita (current US$)
indicator = 'NY.GDP.PCAP.CD'
```

```python
# Set endpoint for GDP per capita, only United States
endpoint = f'https://api.worldbank.org/v2/country/US/indicator/{indicator}?'

# Set params
params ={
    'format' : 'json'
}

# Request to API using GET
response = requests.get(endpoint, params=params)

# Check response
print(response)
usa_data = response.json()
print('Number of records:', len(usa_data[1]))
usa_data
```

2. Adjust your request so that it returns data just for the United States for the year 2021.

```python
# Set endpoint for USA
#endpoint = f'https://api.worldbank.org/v2/country/US/indicator/{indicator}?'

# Set params for 2021
params = {
    'format': 'json',
    'date': '2021'
}

# Request to API
response = requests.get(endpoint, params=params)
usa_2021_data = response.json()

print(response)
print('Number of records for 2021:', len(usa_2021_data[1]))
usa_2021_data
```

3. Adjust your request so that it returns data just for the United States for the years 2000 through 2021.

```python
# Set endpoint
#endpoint = f'https://api.worldbank.org/v2/country/US/indicator/{indicator}?'

# Set params for 2000-2021
params = {
    'format': 'json',
    'date': '2000:2021'
}

# Request to API
response = requests.get(endpoint, params=params)
usa_2000_2021_data = response.json()

print(response)
print('Number of records 2000-2021:', len(usa_2000_2021_data[1]))
usa_2000_2021_data
```

4. Adjust your request so that it returns data for the United States and Canada for the years 2000 through 2021.

```python
# Set endpoint for US and Canada
endpoint = f'https://api.worldbank.org/v2/country/US;CA/indicator/{indicator}?'

# Set params for 20002021
params = {
    'format': 'json',
    'date': '2000:2021'
}

# Request to API
response = requests.get(endpoint, params=params)
us_can_data = response.json()

print(response)
print('Number of records US + Canada:', len(us_can_data[1]))
us_can_data
```
